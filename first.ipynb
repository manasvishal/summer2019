```python
from math import *
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

l=5e6
R=1.5e8
dim=1*365
alp=l/(2*R)
ohm=2*pi/365. 
e= sqrt(1 + 2*alp/sqrt(3) + 4*(alp**2)/3.) - 1
eps=atan(alp/(1 + (alp/sqrt(3))))

t=[i for i in range(dim)]

psi1=[0.0]*dim

for i in range(dim-1):
	psi1[i+1]=ohm*t[i] - e*sin(psi1[i])

X1=[R*(cos(psi1[i])+e)*cos(eps) for i in range(dim)]
Y1=[R*(sqrt(1-eps**2))*sin(psi1[i]) for i in range(dim)]
Z1=[R*(cos(psi1[i])+e)*sin(eps) for i in range(dim)]



############for SC 2#################


psi2=[0.0]*dim

for i in range(dim-1):
	psi2[i+1]=ohm*t[i] - e*sin(psi2[i])	 - 2*pi/3

# X12=[R*(cos(psi2[i])+e)*cos(eps) for i in range(dim)]
# Y12=[R*(sqrt(1-eps**2))*sin(psi2[i]) for i in range(dim)]
# Z12=[R*(cos(psi2[i])+e)*sin(eps) for i in range(dim)]

X2=[(X1[i]*cos(2*pi/3)  -  Y1[i]*sin(2*pi/3)) for i in range(dim)]
Y2=[(X1[i]*sin(2*pi/3)  +  Y1[i]*cos(2*pi/3)) for i in range(dim)]
Z2=[R*(cos(psi2[i] + e))*sin(eps) for i in range(dim)]



############for SC 3#################

psi3=[0.0]*dim

for i in range(dim-1):
	psi3[i+1]=ohm*t[i] - e*sin(psi3[i])	 - 4*pi/3


# X13=[R*(cos(psi3[i])+e)*cos(eps) for i in range(dim)]
# Y13=[R*(sqrt(1-eps**2))*sin(psi3[i]) for i in range(dim)]
# Z13=[R*(cos(psi3[i])+e)*sin(eps) for i in range(dim)]

X3=[(X1[i]*cos(4*pi/3)  -  Y1[i]*sin(4*pi/3)) for i in range(dim)]
Y3=[(X1[i]*sin(4*pi/3)  +  Y1[i]*cos(4*pi/3)) for i in range(dim)]
Z3=[R*(cos(psi3[i] + e))*sin(eps) for i in range(dim)]



d12=[sqrt( (X1[i]-X2[i])**2  + (Y1[i]-Y2[i])**2 + (Z1[i]-Z2[i])**2  ) for i in range(dim)]

d23=[sqrt( (X2[i]-X3[i])**2  + (Y2[i]-Y3[i])**2 + (Z2[i]-Z3[i])**2  ) for i in range(dim)]

d13=[sqrt( (X1[i]-X3[i])**2  + (Y1[i]-Y3[i])**2 + (Z1[i]-Z3[i])**2  ) for i in range(dim)]







plt.plot(X1[3:5],Y1[3:5],Z1[3:5],label="Sat1")
plt.plot(X2[3:5],Y2[3:5],Z2[3:5],label="Sat2")
plt.plot(X3[3:5],Y3[3:5],Z3[3:5],label="Sat3")

plt.legend()
plt.show()

print(d12)
plt.plot(t,d12,label="L12")
plt.hlines(l,0,365,label="5e6 line")
plt.plot(t,d23,label="L23")
plt.plot(t,d13,label="L13")
#plt.ylim(4.8e6,5.1e6)
plt.legend()
plt.show()
















```
